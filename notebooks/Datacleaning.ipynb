# **Electric Vehicle Charge and Range Analysis**

```python
import pandas as pd

evindia = pd.read_csv("EVIndia.csv")
charging = pd.read_csv("electric_vehicle_charging_station_list.csv")
electric_clean = pd.read_csv("ElectricCarData_Clean.csv")
cheapest = pd.read_csv("Cheapestelectriccars-EVDatabase.csv")
```

```python
evindia.info()
evindia.describe()
evindia.isnull().sum()
```

```python
evindia.duplicated().sum()
charging.duplicated().sum()
electric_clean.duplicated().sum()
cheapest.duplicated().sum()
```

```python
electric_clean["Range_Km"].describe()
```

```python
cheapest["PriceinGermany"] = cheapest["PriceinGermany"].replace('[€,]', '', regex=True)
cheapest["PriceinGermany"] = cheapest["PriceinGermany"].astype(float)
```

```python
cheapest["Range"] = pd.to_numeric(cheapest["Range"], errors='coerce')
```

```python
evindia = evindia.drop_duplicates()
```

```python
evindia.to_csv("EVIndia_cleaned.csv", index=False)
charging.to_csv("Charging_cleaned.csv", index=False)
electric_clean.to_csv("ElectricCar_cleaned.csv", index=False)
cheapest.to_csv("CheapestEV_cleaned.csv", index=False)
```

```python
cheapest_new["Range"] = cheapest_new["Range"].astype(str)
cheapest_new["Range"] = cheapest_new["Range"].str.replace(" km","")
cheapest_new["Range"] = pd.to_numeric(cheapest_new["Range"], errors='coerce')
```

```python
cheapest_new["PriceinUK"] = cheapest_new["PriceinUK"].str.replace("£","")
cheapest_new["PriceinUK"] = cheapest_new["PriceinUK"].str.replace(",","")
cheapest_new["PriceinUK"] = pd.to_numeric(cheapest_new["PriceinUK"], errors='coerce')
```

```python
cheapest_new.isnull().sum()
cheapest_new.describe()
```

```python
cheapest_new = cheapest_new.drop(columns=["Range"])
```

```python
cheapest_new["Acceleration"] = cheapest_new["Acceleration"].str.replace(" sec","")
cheapest_new["Acceleration"] = pd.to_numeric(cheapest_new["Acceleration"], errors='coerce')

cheapest_new["TopSpeed"] = cheapest_new["TopSpeed"].str.replace(" km/h","")
cheapest_new["TopSpeed"] = pd.to_numeric(cheapest_new["TopSpeed"], errors='coerce')

cheapest_new["Efficiency"] = cheapest_new["Efficiency"].str.replace(" Wh/km","")
cheapest_new["Efficiency"] = pd.to_numeric(cheapest_new["Efficiency"], errors='coerce')

cheapest_new["FastChargeSpeed"] = cheapest_new["FastChargeSpeed"].str.replace(" km/h","")
cheapest_new["FastChargeSpeed"] = pd.to_numeric(cheapest_new["FastChargeSpeed"], errors='coerce')
```

```python
cheapest_new.info()
cheapest_new.isnull().sum()
cheapest_new.describe()
```
